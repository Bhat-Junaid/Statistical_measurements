#### 1. Probability distribution function of a coin-toss repeated 20 times, getting 12 "heads" and 8 "tails".

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np


N = 20  # Number of bernoulli trails    
p = 0.5 # probability of success
x = np.arange(0, N+1) # Number of successful outcomes i.e heads

# Function to calculate the pdf
def bi_dist(N, p):
    probabilities = [stats.binom.pmf(k, N, p) for k in x]
    return probabilities
prob = bi_dist(N, p)

print("Probability of the various events: ")
print(prob)

# Plotting the pdf
plt.plot(x, prob, label = "pdf", marker = "o")
plt.title(f"Binomial Distribution (N={N}, p={p})")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.xticks(x)
plt.legend()
plt.grid(alpha =  0.5)
plt.show()

# probability of getting 12 heads and 8 tails
print(f"Probability of getting 12 heads: {prob[12]}")
```

####  2: Test of hypothesis by choosing a result as far or farther from the expected result.

```python
# List to store probabilities
sum_prob = []

# calculate the probability of getting a value as far away from the expected value
for i in range (len(prob)):
    if (i <= 8) or (i >= 12):   # less than 8 or greater than 12 heads
        sum_prob.append(prob[i])
        
print(sum(sum_prob))
```

We see that the obtained value is greater than 0.05 which means, we fail to reject the hypothesis that the coin is fair.

|  "Tails"   |  08  |  0.4   |  10   |  0.4  |
|  Sum       |  20  |  1     |  20   |  0.8  |

The final last element of the table(i.e 0.8) represents the chi-square statistic, based on which we can either accept the null hypothesis or reject it. In the upcoming step, we will make the exact calculation of the p-value and make a decision about the null hypothesis based on the calculated p-value. 


#### Calculation of  $\chi^2 $ value and p-value

```python
f_obs = np.array([12, 8])    # observed value of the outcomes
f_exp = np.array([10, 10])   # expected value of the outcomes

# chi-square and p-value calculation
chi_sq_value = stats.chisquare(f_obs, f_exp, ddof = 0)
print(chi_sq_value)
```

$\chi^{2}-statistic = 0.8 $, $p-value = 0.37$

As seen from the above calculations the p-value > 0.05, therefore the hypothesis that the coin is fair can't be rejected.
